```python
import numpy as np
cantSizes = 10;
```

```python
mergeMean = np.zeros(cantSizes)
mergeWorst = np.zeros(cantSizes)
mergeBest = np.zeros(cantSizes)

insertionMean = np.zeros(cantSizes)
insertionWorst = np.zeros(cantSizes)
insertionBest = np.zeros(cantSizes)

selectionMean = np.zeros(cantSizes)
selectionWorst = np.zeros(cantSizes)
selectionBest = np.zeros(cantSizes)

sizes = np.zeros(cantSizes)
```

```python
def readFrom(contenido, beg):
    # Iterar a través del contenido carácter por carácter
    save=0;
    for i in range(beg, len(contenido)):
        # Verificar si el carácter actual es un dígito o un punto decimal
        if contenido[i].isdigit() or contenido[i] == '.':
            # Si es así, construir una cadena que representa el número
            num_str = contenido[i]
            for j in range(i + 1, len(contenido)):
                if contenido[j].isdigit() or contenido[j] == '.':
                    num_str += contenido[j]
                else:
                    save = j 
                    break
            # Convertir la cadena a un valor de punto flotante y salir del bucle
            num = float(num_str)
            #print("El próximo número flotante es:", num)
            return [num, save];
            
    
    
```

```python
f = open('cienmil.txt','r')
contenido = f.read()

idx = 0
for i in range(cantSizes):
    sizes[i], idx = readFrom(contenido,idx)
    
    mergeMean[i], idx = readFrom(contenido,idx)
    mergeWorst[i], idx = readFrom(contenido,idx)
    mergeBest[i], idx = readFrom(contenido,idx)
    
    insertionMean[i], idx = readFrom(contenido,idx)
    insertionWorst[i], idx = readFrom(contenido,idx)
    insertionBest[i], idx = readFrom(contenido,idx)
    
    selectionMean[i], idx = readFrom(contenido,idx)
    selectionWorst[i], idx = readFrom(contenido,idx)
    selectionBest[i], idx = readFrom(contenido,idx)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(sizes, mergeMean, 'o-', label='Merge Mean')
plt.plot(sizes, insertionMean, 's-', label='Insertion Mean')
plt.plot(sizes, selectionMean, 's-', label='Selection Mean')
plt.xlabel('Sizes')
plt.ylabel('Time')
plt.title('Mean case Execution Time')
plt.legend()
plt.show()
```

```python
plt.plot(sizes, mergeWorst, 's-', label='Merge Worst')
plt.plot(sizes, insertionWorst, 's-', label='Insertion Worst')
plt.plot(sizes, selectionWorst, 's-', label='Selection Worst')
plt.xlabel('Sizes')
plt.ylabel('Time')
plt.title('Worst case Execution Time')
plt.legend()
plt.show()
```

```python
plt.plot(sizes, mergeBest, 's-', label='Merge Berst')
plt.plot(sizes, insertionBest, 'o-', label='Insertion Best')
plt.plot(sizes, selectionBest, 's-', label='Selection Best')
plt.xlabel('Sizes')
plt.ylabel('Time')
plt.title('Worst case Execution Time')
plt.legend()
plt.show()
```
